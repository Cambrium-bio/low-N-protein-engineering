## Merge Twist Platemap of all ASR, Consensus, eUniRep designs with metadata to facilitate downstream analysis

```python
import sys
import os

import numpy as np
import pandas as pd
```

```python
plate_mat = pd.read_csv('Twist_plate_map_pSHPs1104B344730AI.csv')
agg_md = pd.read_csv('A009c_GFP_aggregated_unirep_designs_ASR_consensus_seqs.csv')

plate_mat.rename(columns={'Name': 'order_id'}, inplace=True)
```

```python
merge_df = pd.merge(agg_md, plate_mat, on='order_id')

print(merge_df.shape)
merge_df.head()
```

```python
merge_df.to_csv('A009d_metadata_platemap_merged.csv')
```
